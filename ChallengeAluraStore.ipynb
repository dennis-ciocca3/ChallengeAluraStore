## **Importación de Datos**



```python
import pandas as pd
import matplotlib.pyplot as plt

#importando DataFrame
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#almacenando DataFrame es una variable
tienda_1 = pd.read_csv(url1)
tienda_2 = pd.read_csv(url2)
tienda_3 = pd.read_csv(url3)
tienda_4 = pd.read_csv(url4)

#verificando si funciona
tienda_1.head()
```

## **Análisis de Datos**

### 1. Ingreso total por cada tienda



```python
# Ingreso total por cada tienda
ingreso_t1 = tienda_1["Precio"].sum()
ingreso_t2 = tienda_2["Precio"].sum()
ingreso_t3 = tienda_3["Precio"].sum()
ingreso_t4 = tienda_4["Precio"].sum()
```

```python
print(f"El ingreso total de la Tienda 1 es: {ingreso_t1}")
print(f"El ingreso total de la Tienda 2 es: {ingreso_t2}")
print(f"El ingreso total de la Tienda 3 es: {ingreso_t3}")
print(f"El ingreso total de la Tienda 4 es: {ingreso_t4}")
```

### 2. Ventas por categoría

```python
ventas_t1 = tienda_1["Categoría del Producto"].value_counts()
ventas_t2 = tienda_2["Categoría del Producto"].value_counts()
ventas_t3 = tienda_3["Categoría del Producto"].value_counts()
ventas_t4 = tienda_4["Categoría del Producto"].value_counts()
```

```python
print(f"Ventas por categoría - Tienda 1:\n{ventas_t1}\n")
print(f"Ventas por categoría - Tienda 2:\n{ventas_t2}\n")
print(f"Ventas por categoría - Tienda 3:\n{ventas_t3}\n")
print(f"Ventas por categoría - Tienda 4:\n{ventas_t4}")
```

### 3. Calificación promedio de la tienda


```python
valoracion_t1 = tienda_1["Calificación"].mean()
valoracion_t2 = tienda_2["Calificación"].mean()
valoracion_t3 = tienda_3["Calificación"].mean()
valoracion_t4 = tienda_4["Calificación"].mean()
```

```python
print(f"Valoración promedio - Tienda 1: {valoracion_t1}\n")
print(f"Valoración promedio - Tienda 2: {valoracion_t2}\n")
print(f"Valoración promedio - Tienda 3: {valoracion_t3}\n")
print(f"Valoración promedio - Tienda 4: {valoracion_t4}")
```

### 4. Productos más y menos vendidos

```python
mas_vendidos_t1 = tienda_1["Producto"].value_counts().head(5)
menos_vendidos_t1 = tienda_1["Producto"].value_counts().tail(5)

mas_vendidos_t2 = tienda_2["Producto"].value_counts().head(5)
menos_vendidos_t2 = tienda_2["Producto"].value_counts().tail(5)

mas_vendidos_t3 = tienda_3["Producto"].value_counts().head(5)
menos_vendidos_t3 = tienda_3["Producto"].value_counts().tail(5)

mas_vendidos_t4 = tienda_4["Producto"].value_counts().head(5)
menos_vendidos_t4 = tienda_4["Producto"].value_counts().tail(5)
```

```python
print(f"Top 5 productos más vendidos - Tienda 1:\n{mas_vendidos_t1}\n")
print(f"Top 5 productos menos vendidos - Tienda 1:\n{menos_vendidos_t1}\n")

print('-------------------------------------------')
print(f"Top 5 productos más vendidos - Tienda 2:\n{mas_vendidos_t2}\n")
print(f"Top 5 productos menos vendidos - Tienda 2:\n{menos_vendidos_t2}\n")

print('-------------------------------------------')
print(f"Top 5 productos más vendidos - Tienda 3:\n{mas_vendidos_t3}\n")
print(f"Top 5 productos menos vendidos - Tienda 3:\n{menos_vendidos_t3}\n")

print('-------------------------------------------')
print(f"Top 5 productos más vendidos - Tienda 4:\n{mas_vendidos_t4}\n")
print(f"Top 5 productos menos vendidos - Tienda 4:\n{menos_vendidos_t4}")
```

### 5. Envío promedio por tienda

```python
envio_t1 = tienda_1["Costo de envío"].mean()
envio_t2 = tienda_2["Costo de envío"].mean()
envio_t3 = tienda_3["Costo de envío"].mean()
envio_t4 = tienda_4["Costo de envío"].mean()
```

```python
print(f"Costo de envío promedio - Tienda 1: {round(envio_t1, 2)}\n")
print(f"Costo de envío promedio - Tienda 2: {round(envio_t2, 2)}\n")
print(f"Costo de envío promedio - Tienda 3: {round(envio_t3, 2)}\n")
print(f"Costo de envío promedio - Tienda 4: {round(envio_t4, 2)}")

```

## **Visualización**

```python
# TIENDAS

# Datos
tiendas = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]

# Colores
colores = ['#5DA5DA', '#60BD68', '#FAA43A', '#F17CB0']

# Creando gráfico
plt.figure(figsize=(7,5))
plt.bar(tiendas, ingresos, color=colores, edgecolor='black', width=0.6)

# Títulos y etiquetas
plt.title("Ingreso total por tienda", fontsize=14, fontweight='bold')
plt.xlabel("Tiendas", fontsize=12)
plt.ylabel("Ingreso Total", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar valores arriba de cada barra
for i, valor in enumerate(ingresos):
    plt.text(i, valor + (max(ingresos)*0.02), f'{valor:,.0f}',
             ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
print("---------------------------------------------------------------------------------------------------------")

# TIENDA 1
plt.figure(figsize=(8,5))
ventas_t1.plot(kind='bar', color=['#5DA5DA', '#60BD68', '#F15854', '#FAA43A', '#B276B2'], edgecolor='black')
plt.title("Distribución de categorías - Tienda 1", fontsize=14, fontweight='bold')
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(8,5))
plt.plot(tienda_1["Precio"], tienda_1["Costo de envío"],
         color='#4C72B0', linewidth=2)  # línea simple sin adornos
plt.title("Precio vs Costo de Envío - Tienda 1", fontsize=14, fontweight='bold')
plt.xlabel("Precio", fontsize=12)
plt.ylabel("Costo de Envío", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
print("---------------------------------------------------------------------------------------------------------")

# TIENDA 2
plt.figure(figsize=(8,5))
ventas_t2.plot(kind='bar', color=['#60BD68', '#F15854', '#B276B2', '#FAA43A', '#5DA5DA'], edgecolor='black')
plt.title("Distribución de categorías - Tienda 2", fontsize=14, fontweight='bold')
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(8,5))
plt.plot(tienda_2["Precio"], tienda_2["Costo de envío"],
         color='#55A868', linewidth=2)  # línea simple sin adornos
plt.title("Precio vs Costo de Envío - Tienda 2", fontsize=14, fontweight='bold')
plt.xlabel("Precio", fontsize=12)
plt.ylabel("Costo de Envío", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
print("---------------------------------------------------------------------------------------------------------")

# TIENDA 3
plt.figure(figsize=(8,5))
ventas_t3.plot(kind='bar', color=['#F15854', '#5DA5DA', '#60BD68', '#FAA43A', '#B276B2'], edgecolor='black')
plt.title("Distribución de categorías - Tienda 3", fontsize=14, fontweight='bold')
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(8,5))
plt.plot(tienda_3["Precio"], tienda_3["Costo de envío"],
         color='#C44E52', linewidth=2)  # línea simple sin adornos
plt.title("Precio vs Costo de Envío - Tienda 3", fontsize=14, fontweight='bold')
plt.xlabel("Precio", fontsize=12)
plt.ylabel("Costo de Envío", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
print("---------------------------------------------------------------------------------------------------------")

# TIENDA 4
plt.figure(figsize=(8,5))
ventas_t4.plot(kind='bar', color=['#B276B2', '#60BD68', '#F15854', '#FAA43A', '#5DA5DA'], edgecolor='black')
plt.title("Distribución de categorías - Tienda 4", fontsize=14, fontweight='bold')
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Cantidad de productos vendidos", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

print()
plt.figure(figsize=(8,5))
plt.plot(tienda_4["Precio"], tienda_4["Costo de envío"],
         color='#8172B2', linewidth=2)  # línea simple sin adornos
plt.title("Precio vs Costo de Envío - Tienda 4", fontsize=14, fontweight='bold')
plt.xlabel("Precio", fontsize=12)
plt.ylabel("Costo de Envío", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
